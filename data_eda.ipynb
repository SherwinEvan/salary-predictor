```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Removing rows with broken data

cols = ['Location', 'Size', 'Founded', 'Type of ownership', 'Industry', 'Sector']
for col in cols:
    df = df.drop(df[df[col].str.contains('USD')].index)
```

```python
# Function to classify common Job Titles

def title_simplifier(title):
    if 'full stack developer' in title.lower():
        return 'full stack developer'
    elif 'full stack engineer' in title.lower():
        return 'full stack engineer'
    elif 'web developer' in title.lower():
        return 'web developer'
    elif 'mern' in title.lower():
        return 'mern'
    elif 'mean' in title.lower():
        return 'mean'
    elif '.net' in title.lower():
        return '.net'
    elif 'software developer' in title.lower():
        return 'software developer'
    elif 'software engineer' in title.lower():
        return 'software engineer'
    elif 'full stack' in title.lower():
        return 'full stack'
    elif 'backend' in title.lower():
        return 'backend'
    elif 'frontend' in title.lower():
        return 'frontend'
    else:
        return 'na'
```

```python
# Function to classify Seniority level

def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
df['job_simp'] = df['Job Title'].apply(title_simplifier)
df.job_simp.value_counts()
```

```python
df['Seniority'] = df['Job Title'].apply(seniority)
df['Seniority'].value_counts()
```

```python
df['desc_len'] = df['Job Description'].apply(lambda x: len(x))
df['desc_len']
```

```python
df.columns
```

```python
df.rename(columns = {'Seniority':'seniority'}, inplace = True)
```

```python
#Monthly wage to annual 

df['Min Salary'] = df.apply(lambda x: x['Min Salary'] * 0.12 if x['Monthly'] == 1 else x['Min Salary'], axis = 1)
df['Max Salary'] = df.apply(lambda x: x['Max Salary'] * 0.12 if x['Monthly'] == 1 else x['Max Salary'], axis = 1)
df['Avg Salary'] = df.apply(lambda x: (x['Max Salary'] + x['Min Salary']) / 2 if x['Monthly'] == 1 else x['Avg Salary'], axis = 1)
```

```python
df[df['Monthly'] == 1][['Monthly', 'Min Salary', 'Max Salary', 'Avg Salary']]
```

```python
#Hourly wage to annual 

df['Min Salary'] = df.apply(lambda x: x['Min Salary'] * 0.01440 if x['Per Hour'] == 1 else x['Min Salary'], axis = 1)
df['Max Salary'] = df.apply(lambda x: x['Max Salary'] * 0.01440 if x['Per Hour'] == 1 else x['Max Salary'], axis = 1)
df['Avg Salary'] = df.apply(lambda x: (x['Max Salary'] + x['Min Salary']) / 2 if x['Per Hour'] == 1 else x['Avg Salary'], axis = 1)
```

```python
df[df['Per Hour'] == 1][['Per Hour', 'Min Salary', 'Max Salary', 'Avg Salary']]
```

```python
#Cleaning escape sequence from Company name

df['Company'] = df['Company'].apply(lambda x: x.replace('\r\n', ''))
df['Company']
```
